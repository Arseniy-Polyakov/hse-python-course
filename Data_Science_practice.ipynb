Как читать ящик с усами?

- Полоска слева - нижняя граница (минимум)
- Две части ящика разделяет медиана
- Полоска справа - верхняя граница (максимум)
- Точки - выбросы

Наш ящик расположен по горизонтали. Чаще ящики с усами рисуют по вертикали. Расположение не имеет принципиального значения.

# Группировка данных

````python
# Коррелограмма отображает только числовые связи,
# но некоторые данные в датасете категориальные, то есть они описаны словами
# Посмотрим, как тип топлива влияет на стоимость автомобиля
# Группируем категориальный столбец с числовым и применяем фукнцию для подсчета среднего значения
grouped_data = df.groupby('### столбец с типом топлива ###')['### столбец со стоимостью ###'].mean()

grouped_data
````
