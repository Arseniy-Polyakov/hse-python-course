**`Matplotlib`**: библиотека для отрисовки графиков, созданных с помощью таких библиотек, как `Numpy` и `Pandas`.

**Сравнение `Matplotlib` и `Seaborn`**

`Matplotlib`|`Seaborn`
------------|--------
Простые графики: столбчатые, круговые, точечные диаграммы|Beautification; дистрибуция и связи, коррелограммы
Образец синтаксиса: `matplotlib.pyplot.bar(x_axis, y_axis)`|Образец синтаксиса: `seaborn.barplot(x_axis, y_axis)`
Построен на базе `MATLAB`, аналогичные фичи и синтаксис|Высокая совметимость с `Pandas`, датасет как единица
